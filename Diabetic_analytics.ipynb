```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data present in the TSV file
path="/content/drive/MyDrive/diabetese.tsv"
df= pd.read_csv(path,sep="\t")
df
```

```python
df.head()
```

```python
df.info()
```

```python
#let's find the average age of paitents
mean_age=df['AGE'].mean()
print(f" average age of paitents is : {mean_age}")
```

```python
# Variance
var_age=df["AGE"].var()
print (f"variance of age is : {var_age}")
```

```python
std_age=df["AGE"].std()
print(f"standard deviation is :{std_age}")
```

```python
#Boxplot for the age field
plt.figure(figsize=(2,6))
plt.boxplot(df["AGE"],showmeans=True)
plt.grid(color="grey",linestyle="dotted")
plt.xlabel("AGE")
plt.ylabel("years")
plt.show()
```

```python
#insights of Boxplot
# 25% of people are less than 38 years of age
# 50% of people are less than 48 years of age
# 75% of people are less than 59 years of age
```

```python
# Boxplot for BMI
plt.figure(figsize=(6,2))
plt.boxplot(df["BMI"],showmeans=True, vert=False)
plt.grid(color="grey",linestyle="dotted")
plt.xlabel("Value")
plt.ylabel("BMI")
plt.show()
```

```python
# Boxplot for BP
plt.figure(figsize=(6,2))
plt.boxplot(df["BP"],showmeans=True, vert=False)
plt.grid(color="grey",linestyle="dotted")
plt.xlabel("Value")
plt.ylabel("BP")
plt.show()
```

```python
# Histogram = frequency
# what age of paitents are mostly diabetic
df["AGE"].hist(bins=12) # bins means range
plt.title("age of distributation of the diabetic paitents")
plt.xlabel("age range")
plt.ylabel("frequency(count)")
plt.show()
```

```python
#insight of histplot
# age range 49 to 55 are max people are diabetic
```

```python
df["BMI"].hist(bins=12) # bins means range
plt.title("BMI of distributation of the diabetic paitents")
plt.xlabel("BMIrange")
plt.ylabel("frequency(count)")
plt.show()
```

```python
#scatter plot
plt.scatter(df["Y"],df["BMI"])
plt.xlabel("disease progression")
plt.ylabel("BMI")
plt.title("Scatter plot")
plt.show()
```

```python
plt.scatter(df["Y"],df["S5"])
plt.xlabel("disease progression")
plt.ylabel("S5 bolld type")
plt.title("Scatter plot : S5 Vs Disease progression ")
plt.show()
```
